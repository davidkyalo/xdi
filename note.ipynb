```python
from xdi.containers import Container
import networkx as nx
import matplotlib.pyplot as plt

a = Container('A')
b = Container('B')
c = Container('C')
d = Container('D')
e = Container('E')
f = Container('F')
g = Container('G')

G = nx.DiGraph()
```

```python
a.include(b, c)
b.include(d)
c.include(e)

d.include(f)
e.include(g)


# G.add_nodes_from([a, b, c, d, e, f, g])

G.add_edges_from([
    (a, a), (b, b), (c, c), (d, d), (e, e), (f, f), (g, g),
    (a, b), (a, c),
    (b, d), (c, e),
    (d, e),
    (d, f), (e, g)
][::-1])

fig = plt.figure(1, figsize=(8
, 6))

kwds = dict(
    with_labels=True,
    arrowsize=24,
    node_size=400,
    # font_size=14,

)

# nx.draw(G, **kwds)
nx.draw_circular(G, **kwds)

# nx.draw_spring(G, with_labels=True, node_size=60,font_size=14)
# G
```
